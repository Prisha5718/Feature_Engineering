Feature Engineering

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score
```

```python
x=df.drop(["quality","quality_Label"],axis=1)
y=df["quality_Label"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
Scalar=StandardScaler()
x_train_scaled=Scalar.fit_transform(x_train)
x_test_scaled=Scalar.transform(x_test)

baseline_model=LogisticRegression()
baseline_model.fit(x_train_scaled,y_train)

y_pred_baseLine=baseline_model.predict(x_test_scaled)
```

```python
print("BaseLine Accuracy: ",accuracy_score(y_test,y_pred_baseLine))
print("BaseLine Precision: ",precision_score(y_test,y_pred_baseLine))
print("BaseLine Recall: ",recall_score(y_test,y_pred_baseLine))
```

Making Features

Combined Acidity - Feature 1

```python
df["Total_Acidity"]=(
    df["fixed acidity"]+
    df["volatile acidity"]+
    df["citric acid"]
)
```

Sulfur Ratio - Feature 2

```python
df["Sulfur_Ratio"]=(
    df["free sulfur dioxide"]/(df["total sulfur dioxide"]+1)
)
```

Residual_Sugar_Ratio

```python
df["Residual_Sugar_Ratio"]=(
    df["residual sugar"]/(df["density"]+1)
)
```
